```python
import pandas as pd
import altair as alt
```

```python
df_eva = pd.read_csv('C:/Users/ilove/OneDrive/Uni/Master - Philipps Uni/3. Semester/Seminar - Lakehouse/Code/evaluation_dataset.csv')
```

```python
null_0 = (df_eva['Null difference'] == 0).sum() # Anzahl Tabellen wo keine Werte ersetzt wurden
null_1 = (df_eva['Null difference'] != 0).sum() # ANzahl Tabellen wo Werte ersetzt wurden
null_overall = null_0 + null_1 # Anzahl Tabellen mit Null-Werten
null_befor_overall = df_eva['Null befor'].sum() # Anzahl Null-Values insgesamt
null_replaced_ovrall = df_eva['Null difference'].sum() # Anzahl ersetzter Null-Values insgesamt
null_replaced_mean = null_replaced_ovrall / null_overall # Anzahl durchschnittlich ersetzter Werte pro Tabelle
replacemnt_rate_overall = (null_replaced_ovrall / null_befor_overall) * 100 # Replacement Rate insgesamt

print('Tables with Null-Values and FDs:', len(df_eva))
print('Tables with Replacement Rate = 0:', null_0)
print('Tables with Replacement Rate > 0:', null_1)
print('Null-Values befor:', null_befor_overall)
print('Replaced Null-Values:', null_replaced_ovrall)
print('Mean number of replaced Null-Values:',round(null_replaced_mean, 2))
print('Replacement Rate Overall:', round(replacemnt_rate_overall, 2))
```

## Overview Rate Distribution

```python
df_eva_without_0 = df_eva.loc[df_eva['replacement rate'] != 0]
df_eva_g_10 = df_eva.loc[(df_eva['replacement rate'] >= 10)]
df_eva_k_10 = df_eva.loc[(df_eva['replacement rate'] != 0) & (df_eva['replacement rate'] < 10)]

rate_over_view_aa = alt.Chart(df_eva_without_0).mark_bar(filled=True).encode(
    alt.X('replacement rate', bin = True, axis=alt.Axis(title='Replacement Rate per Table')),
    alt.Y ('count()', axis=alt.Axis(title='Number of Tables'))
).properties(width=400, height=300, title = 'Replacement Rate Value Distribution (overview)')

rate_over_view_ab1 = alt.Chart(df_eva_k_10).mark_bar(filled=True).encode(
    alt.X('replacement rate', bin = True, axis=alt.Axis(title='Replacement Rate per Table')),
    alt.Y ('count()', axis=alt.Axis(title='Number of Tables'))
).properties(width=400, height=300, title = 'Replacement Rate Value Distribution (Rate < 10)')

rate_over_view_ab2 = alt.Chart(df_eva_g_10).mark_bar(filled=True).encode(
    alt.X('replacement rate', bin = True, axis=alt.Axis(title='Replacement Rate per Table')),
    alt.Y ('count()', axis=alt.Axis(title='Number of Tables'))
).properties(width=400, height=300, title = 'Replacement Rate Value Distribution (Rate > 10)')

rate_over_view_aa & (rate_over_view_ab1 | rate_over_view_ab2)
```

```python
df_eva_90 = df_eva.loc[df_eva['replacement rate'] >= 90] # Tables with a replacement rate > 90%
null_befor_90 = df_eva_90['Null befor'].sum() # number of Null values befor
null_after_90 = df_eva_90['Null after'].sum() # numer of Null values after
mean_rate_90 = null_befor_90 / len(df_eva_90)

df_eva_3 = df_eva.loc[(df_eva['replacement rate'] <= 3.01) & (df_eva['Null difference'] != 0)] # Tables with a replacement rate > 90%
null_befor_3 = df_eva_3['Null befor'].sum() # number of Null values befor
null_after_3 = df_eva_3['Null after'].sum() # numer of Null values after
mean_rate_3 = null_befor_3 / len(df_eva_3)

print("Number of tables with a rate < 3%:", len(df_eva_3))
# print("Number of Null values befor:", null_befor_3)
# print("Number of Null values after:", null_after_3)
print("Mean Number of Null values per Table (Rate <= 3.01%):", round(mean_rate_3, 2))
print('----------------------------------------------------------------------')
print("Number of tables with a rate > 3.01%:", len(df_eva_90))
# print("Number of Null values befor:", null_befor_90)
# print("Number of Null values after:", null_after_90)
print("Mean Number of Null values per Table (Rate >= 90%):", round(mean_rate_90, 2))
```

## Replaces Null-Values

```python
df_eva_without_0 = df_eva.loc[df_eva['replacement rate'] != 0]
df_eva_g_50 = df_eva.loc[(df_eva['Null difference'] >= 50)]
df_eva_k_50 = df_eva.loc[(df_eva['Null difference'] != 0) & (df_eva['Null difference'] < 50)]

difference_a = alt.Chart(df_eva_without_0).mark_bar(filled=True).encode(
    alt.X('Null difference', bin=True, axis=alt.Axis(title='Number of replaced Null values per Table')),
     alt.Y ('count()', axis=alt.Axis(title='Number of Tables'))
).properties(width=400, height=300, title = 'Number of replaced Null values (Overview)')

difference_ba = alt.Chart(df_eva_k_50).mark_bar(filled=True).encode(
    alt.X('Null difference', bin=True, axis=alt.Axis(title='Number of replaced Null values per Table')),
    alt.Y ('count()', axis=alt.Axis(title='Number of Tables'))
).properties(width=400, height=300, title = 'Number of replaced Null values (number < 50)')

difference_bb = alt.Chart(df_eva_g_50).mark_bar(filled=True).encode(
    alt.X('Null difference', bin=True, axis=alt.Axis(title='Number of replaced Null values per Table')),
    alt.Y ('count()', axis=alt.Axis(title='Number of Tables'))
).properties(width=400, height=300, title = 'Number of replaced Null values (number > 50)')

difference_a & (difference_ba | difference_bb)
```
